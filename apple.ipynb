```python
ProductNo=[]
Product=[]
Offer=[]
Rating=[]
PastPrice=[] 
PresentPrice=[]
with open('Apples1.txt') as f:
    for i in f:
        if i.startswith('Product No') :
            ProductNo.append(i.split(':')[1].rstrip('\n'))
        elif i.startswith('Product') :
            Product.append(i.split(':')[1].rstrip('\n'))
        elif i.startswith('PresentPrice') :
            PresentPrice.append(i.split(':')[1].rstrip('\n'))
        elif i.startswith('PastPrice'):
            PastPrice.append(i.split(':')[1].rstrip('\n'))
        elif i.startswith('Rating') :
            Rating.append(i.split(':')[1].rstrip('\n'))
        elif i.startswith('Offer') :
            Offer.append(i.split(':')[1].rstrip('\n'))
```

```python
i=1
for j in PresentPrice:
    print(f'{i} :{j}')
    i+=1
```
